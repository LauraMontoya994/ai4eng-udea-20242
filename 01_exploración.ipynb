# Load train dataset (train.csv)

```python
import pandas as pd
df = pd.read_csv('train.csv')
df.head()
```

## General info

```python
print(" "*10 + "Info: \n")
df.info()
print("\n", "-"*30, "\n")
print(" "*10 + "Descripción numérica: \n")
df.describe()
```

```python
print("# Valores únicos:")
df.nunique()
```

```python
print("Valores nulos:")
df.isnull().sum()
```

```python
print("Algunas filas con valores nulos:")
df[df.isnull().any(axis=1)].head()
```

```python
print(f"# de depts: {df['ESTU_PRGM_DEPARTAMENTO'].nunique()} \n")
print("# Estudiantes por dept:")
df.groupby('ESTU_PRGM_DEPARTAMENTO').count()["ID"].sort_values(ascending=False)
```
